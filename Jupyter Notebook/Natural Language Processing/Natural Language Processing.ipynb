```
df = pd.read_csv("labeled_data.csv",sep=",")
df
```

```
tweets = df['tweet']
```

```
import html
import emoji

def replace_emoji(tweet):
  tweet = emoji.demojize(html.unescape(tweet))
  return tweet

tweets = tweets.apply(replace_emoji)
```

```
from nltk import word_tokenize

tokens = tweets
tokens = tokens.apply(word_tokenize)
tokens
```

```
from nltk.corpus import stopwords
en_stopw = set(stopwords.words('english'))

def clean_stopwords(tweet):
  lista = []
  for word in tweet: 
    if word not in en_stopw: 
      lista.append(word)
  return lista

tokens = tokens.apply(clean_stopwords)
tokens
```

```
contractions = {'\'s':'is', '\'ve': 'have', 'n\'t':'not', '\'ll':'will', '\'re': 'are', '\'m': 'am'}


def remove_contractions(tweet):
  lista = []
  for word in tweet:
    if word in contractions:
      lista.append(contractions[word])
    else:
      lista.append(word)

  return lista

tokens = tokens.apply(remove_contractions)
tokens
```

```
unuseful = ['!', '"', '$', '%', '&', '/', '(', ')', '=', '_', '^', '*', '¡', '@', ':', ';', '#', 'RT', '.', '...']

def clean_unuseful(tweet):
  lista = []
  for word in tweet:
    if word not in unuseful:
      lista.append(word)
  return lista

tokens = tokens.apply(clean_unuseful)
tokens
```

```
def remove_capital(tweet):
  lista = []
  for word in tweet:
    lista.append(word.lower())
  return lista

tokens = tokens.apply(remove_capital)
tokens
```

```
# Sacar palabra origen/comun

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

def apply_lemmatize(tweet):
  return [lemmatizer.lemmatize(word) for word in tweet]

tokensLem = tokens.apply(apply_lemmatize)
tokensLem
```

```
# Reducir a raiz

from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

def apply_stemmer(t):
  return [stemmer.stem(word) for word in t]

tokensStem = tokens.apply(apply_stemmer)
tokensStem
```

```
# TFIDF

from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import sent_tokenize
from tqdm.notebook import tqdm

vectorizer = TfidfVectorizer(ngram_range=(1, 2), min_df=1)

vectorized_tweets = []

for tweet in tqdm(tokens):
  X = vectorizer.fit_transform(sent_tokenize(' '.join(tweet)))

  pos_vocab = {v:i for i, v in enumerate(vectorizer.get_feature_names())}

  tfidf = X.toarray()

  vectorized_tweets.append(tfidf)
```
